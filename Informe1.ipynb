## Modelos Cosmológicos

Dado un modelo cosmológico determinado por:

- La constante de Hubble $H_0$
- La densidad de materia $\Omega_m$
- La densidad de radiación $\Omega_r$
- La densidad de curvatura $\Omega_k$
- La densidad de ernegía oscura $\Omega_\Lambda$

Se puede calcular la denominada distancia cosmológica a partir de las siguiente operaciones matemáticas

Definimos la distancia de Hubble:

$$D_H = \dfrac{c}{H_0}$$

Definimos el parámetro de Hubble

$$E(z) = [\Omega_r(1+z)^4 + \Omega_m(1+z)^3 + \Omega_k(1+z)^2 + \Omega_\Lambda]^{1/2}$$

La distancia comoviente puede calcularse como

$$d_c = D_H \int_0^z \dfrac{dz'}{E(z')}$$

Mientras que la distancia Transversa comoviente queda definida como

$$ d_M(z) = \{ \begin{align}
& \frac{d_H}{\sqrt{\Omega_k}} \sinh\left(\sqrt{\Omega_k}d_C(z)/d_H\right) & \text{for } \Omega_k>0\\
& d_C(z) & \text{for }\Omega_k=0\\
& \frac{d_H}{\sqrt{|\Omega_k}|} \sin\left(\sqrt{|\Omega_k|}d_C(z)/d_H\right) & \text{for }\Omega_k<0
\end{align} $$

Finalmente la distancia luminosa puede escribirse como

$$d_L = (1+z) d_M$$

Definimos algunos modelos cosmológicos prototípicos para poder hacer una rápida comparación del efecto de cada parámetro

## Ajuste de los parámetros cosmológicos

A continuanción vamos a definir un modelo para ser ajustado mediante un esquema del tipo Markov Chain Monte Carlo (MCMC)

Como el término de radiación en el factor de Hubble deja de ser dominante a redshifts bajos como los que estamos trabajando vamos a despreciar la densidad de radiación. Si además consideramos un universo plano, nos queda un sólo parámetro cosmológico libre en nuestro modelo: la densidad de materia.

El modelo puede considerar además el ajuste de los parámetros de calibración $M_b$, $\alpha$ y $\beta$. Pero hay que tener en cuenta que el método de las supernovas tipo Ia mide distancias relativas, por ello no es posible ajustar simultaneamente la constante $H_0$ y el valor de $M_b$. 

En principio vamos a asumir priors uniformes para cada uno de los parámetros libres de nuestro modelo. Si deseamos obviar el ajuste de un parámetro por cualquier razón, simplemente podemos dejar el valor de ese parámetro constante durante todo el algoritmo de ajuste.

### Comparación de los modelos con y sin correcciones al modulo de distancia

Realizamos dos ajustes. El primero no tiene en cuenta las correcciones al modulo de distancia, mientras que el segundo asume los valores fiduciales constantes antes mencionados. Puede observarse en la siguiente figura el ajuste con correcciones es significativamente mejor como puede apreciarse a partir de los residuos obtenidos en cada caso.

![caption](files/residuos.jpg)

Los resultados obtenidos fueron los siguientes

|Modelo              | $H_0$ [km s$^{-1}$ Mpc$^{-1}$]  |  $\Omega_m$        |
|--------------------|---------------------------------|--------------------|
|Sin correcciones    | $72.23 \pm 0.26$                |  $0.478 \pm 0.015$ |
|Con correcciones    | $72.90 \pm 0.32$                |  $0.270 \pm 0.017$ |

### Ajuste de los factores de corrección

También realizamos ajustes dejando como parámetros libres los factores de corrección. En este caso tomamos un valor constante de $H_0 = 70 km s^{-1} Mpc^{-1}$ y obtuvimos como resultado:

|Modelo              | $M_b$               | $\alpha$          | $\beta$         | $\Omega_m$      |
|--------------------|---------------------|-------------------|-----------------|-----------------|
|Correcciones libres | $-19.083 \pm 0.009$ | $0.113 \pm 0.006$ | $2.55 \pm 0.07$ | $0.31 \pm 0.02$ |


### Ajuste sin asumir Universo plano

Realizamos un ajuste final relajando además la condición de Universo plano. De esta forma se puede ajustar simultaneameante la densidad de matería y la densidad de energía oscura. Los resultados obtenidos fueron los siguientes:

|Modelo          | $M_b$               | $\alpha$          | $\beta$         | $\Omega_m$      | $\Omega_\Lambda$ |
|----------------|---------------------|-------------------|-----------------|-----------------|------------------|
|Curvatura libre | $-19.08 \pm 0.01$   | $0.113 \pm 0.006$ | $2.55 \pm 0.07$ | $0.28 \pm 0.07$ |$0.66 \pm 0.12$   |


La figura muestra los contornos de probabilidad de 68% y 95% para los parámetros $\Omega_m$ y $\Omega_\Lambda$ mientras que la línea rayada corresponde a los modelos cosmológicos con geometría plana.
